```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
IRIS = load_iris()
iris = pd.DataFrame(data=IRIS.data, columns=IRIS.feature_names)
```

```python
iris
```

```python
iris["class"] = IRIS.target
iris["class"] = iris["class"].map({0:'setosa', 1:'versicolor', 2:'virginica'})
```

```python
iris.groupby(by='class').mean()
```

```python
print(type(iris.groupby(by='class')))
print(type(iris.groupby(by='class').mean()))
```

```python
iris.groupby(by='class').median()
```

```python
pd.Series(iris['class']).value_counts()
```

```python
iris['class'].value_counts()
```

```python
iris.columns = ['sep_eln', 'sep_wid', 'pet_len', 'pet_wid', 'class']
iris
```

```python
#qcut은 quantile 컷으로 q분위수를 기준으로 나눈다
# 절대값으로 자르는 cut()도 존재
iris['petal width level'] = pd.qcut(iris['pet_wid'], q=3, labels=['short', 'middle', 'long'])
iris['petal width level'].value_counts().to_frame()
```
