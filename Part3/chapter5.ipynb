```python
import matplotlib.pyplot as plt

plt.hlines(-6, -10, 10, color='grey', linestyle='solid')
plt.hlines(-2, -10, 10, color='green')
plt.hlines(2, -10, 10, color='orange')
plt.hlines(6, -10, 10, color='red')
plt.vlines(-6, -10, 10, linestyles= 'solid')
plt.vlines(-2, -10, 10, linestyles= 'dashed')
plt.vlines(2, -10, 10, linestyles= 'dashdot')
plt.vlines(6, -10, 10, linestyles= 'dotted')
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris


IRIS = load_iris()
iris = pd.DataFrame(IRIS.data, columns=IRIS.feature_names)
iris["class"] = IRIS.target
iris["class"] = iris["class"].map({0:'Setosa', 1:'Versicolor', 2:'Virginica'})
iris.columns = ['sep_len', 'sep_wid', 'pet_len', 'pet_wid', 'class']
del IRIS
iris
```

```python
linear_func = lambda x: 2*x+1

X = iris['sep_len']
# X
# linear_func(X)
plt.plot(X, linear_func(X), c='#789395')
```

```python
import numpy as np


X, Y = iris['sep_len'], iris['pet_len']
b1, b0 = np.polyfit(X, Y, 1)
plt.scatter(X, Y, alpha=0.5)
plt.plot(X, b1*X + b0, color='red')

plt.show()
```

```python
iris2 = iris.sort_values(by='sep_len')
X, Y = iris2['sep_len'], iris2['pet_len']
b2, b1, b0 = np.polyfit(X, Y, 2)
plt.scatter(X, Y, alpha=0.5)
plt.plot(X, b2*X**2 + b1*X + b0, color='red')

plt.show()
```

```python
plt.plot('sep_len', 'pet_len', data=iris2)
plt.show()
```

```python
# iris2['class'].value_counts()
plt.plot('sep_len', 'pet_len', data = iris2.loc[iris2['class'] == 'Setosa'])
plt.plot('sep_len', 'pet_len', data = iris2.loc[iris2['class'] == 'Versicolor'])
plt.plot('sep_len', 'pet_len', data = iris2.loc[iris2['class'] == 'Virginica'])
```
