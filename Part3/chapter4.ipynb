```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

IRIS = load_iris()
iris = pd.DataFrame(IRIS.data, columns=IRIS.feature_names)
iris["class"] = IRIS.target
iris["class"] = iris["class"].map({0:'Setosa', 1:'Versicolor', 2:'Virginica'})
iris.columns = ['sep_len', 'sep_wid', 'pet_len', 'pet_wid', 'class']
del IRIS
iris
```

```python
plt.title('iris scatter')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')

plt.scatter(x=iris['sep_len'], y=iris['sep_wid'], alpha=0.5)
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(x='sep_len', y='sep_wid', data=iris, hue='class', style='class')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```
