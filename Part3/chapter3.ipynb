```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

IRIS = load_iris()
iris = pd.DataFrame(IRIS.data, columns=IRIS.feature_names)
iris["class"] = IRIS.target
iris["class"] = iris["class"].map({0:'setosa', 1:'versicolor', 2:'virginica'})
iris.columns = ['sep_len', 'sep_wid', 'pet_len', 'pet_wid', 'class']
iris
```

```python
plt.boxplot(iris.drop(columns='class'))
plt.show()
```

```python
plt.boxplot(iris['sep_wid'])
plt.title('Separ Width as cm')
plt.show()
```

```python
iris[['sep_wid', 'class']].boxplot(by='class')
# type(iris[['sep_len', 'class']])
# iris.boxplot(by='class')
```

```python
# 책과 같은 그림이 되기 위해선 hue 변수 필요 누락됨
import seaborn as sns
sns.boxplot(x='class', y='sep_wid', data=iris, hue='class', legend='full')
plt.show()
```
